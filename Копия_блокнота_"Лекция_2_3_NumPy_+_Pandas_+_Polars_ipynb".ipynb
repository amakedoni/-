## Объединение массивов

```python
arr1 = np.array([1, 2, 3])
arr2 = np.array([4, 5, 6])

arr = np.concatenate((arr1, arr2))
print(arr)
```

```python
arr1 = np.array([[1, 2], [3, 4]])
arr2 = np.array([[5, 6], [7, 8]])

arr = np.concatenate((arr1, arr2), axis=0)
print(arr)
print('____________')
arr = np.concatenate((arr1, arr2), axis=1)
print(arr)
```

```python
arr1 = np.array([[1, 2], [3, 4]])
arr2 = np.array([[5, 6, 9], [7, 8, 7]])

arr = np.concatenate((arr1, arr2), axis=0)
print(arr)
print('____________')
arr = np.concatenate((arr1, arr2), axis=1)
print(arr)
```

```python
# stack позволяет объединять по новой оси

arr1 = np.array([1, 2, 3])
arr2 = np.array([4, 5, 6])

arr = np.stack((arr1, arr2), axis=1)
print(arr)
```

```python
# vstack складывает массивы друг под другом (по строкам)

arr1 = np.array([1, 2, 3])
arr2 = np.array([4, 5, 6])

arr = np.vstack((arr1, arr2))
print(arr)

print('_____')

a = np.array([[1, 2], [3, 4]])   # 2 строки
b = np.array([[5, 6]])           # 1 строка

result = np.vstack([a, b])
print(result)
```

```python
# hstack - по горизонтали

a = np.array([1, 2, 3])
b = np.array([4, 5, 6])

result = np.hstack([a, b])

print(result)
print('_________')

a = np.array([[1], [2], [3]])   # 3 строки, 1 столбец
b = np.array([[4], [5], [6]])   # 3 строки, 1 столбец

result = np.hstack([a, b])
print(result)
```

## Работа с пропусками - np.nan

| Характеристика | `None` | `np.nan` |
| :--- | :--- | :--- |
| **Тип данных** | `NoneType` | `float` |
| **Сравнение** | `None == None` → `True` | `np.nan == np.nan` → `False` |
| **Как проверять** | `x is None` | `np.isnan(x)` или `pd.isna(x)` |
| **Арифметика** | Вызывает `TypeError` | Результат становится `nan` |
| **В NumPy массиве** | `dtype('O')` (object) | `dtype('float64')` |
| **Производительность** | Медленно (Python объекты) | Быстро (векторизация C) |
| **Pandas (числа)** | Преобразует столбец в `object` | Сохраняет числовой тип |
| **Сериализация JSON** | Превращается в `null` | Требует специальной обработки |
| **Когда использовать** | Логика Python, строки, объекты | Числовые данные, анализ, матрицы |

```python
print("\n2. ОБНАРУЖЕНИЕ np.nan")
print("-" * 40)

# ❌ НЕПРАВИЛЬНО: nan != nan
print(f"np.nan == np.nan : {np.nan == np.nan}")  # False!

# ✅ ПРАВИЛЬНО: используем специальные функции
print(f"np.isnan(np.nan) : {np.isnan(np.nan)}")          # True
print(f"math.isnan(np.nan) : {math.isnan(np.nan)}")      # True
print(f"pd.isna(np.nan) : {pd.isna(np.nan)}")            # True
print(f"pd.isna(None) : {pd.isna(None)}")                # True (универсально!)

# Для массивов
arr = np.array([1, 2, np.nan, 4, np.nan])
print(f"\nМассив: {arr}")
print(f"np.isnan(arr) : {np.isnan(arr)}")                # [False False  True False  True]
print(f"Количество nan: {np.sum(np.isnan(arr))}")        # 2
```

## Копирование массивов

```python
arr = np.array([1, 2, 3, 4, 5])
x = arr
arr[0] = 42

print(arr)
print(x)
```

```python
arr = np.array([1, 2, 3, 4, 5])
x = arr.copy()
arr[0] = 42

print(arr)
print(x)
```

```python
arr = np.array([1, 2, 3, 4, 5])
x = arr.view()
arr[0] = 42

print(arr)
print(x)
```

```python
arr = np.array([1, 2, 3, 4, 5])
x = arr.view()
x[0] = 42

print(arr)
print(x)
```

| Операция | Результат | Пример | Комментарий |
| :--- | :--- | :--- | :--- |
| **Присваивание** | Ссылка | `b = a` | Переменные указывают на один объект. |
| **Срезы (Slicing)** | **View** | `b = a[1:5]` | По умолчанию возвращает view (без копирования данных). |
| **Изменение формы** | **View** | `b = a.reshape(...)` | Если память непрерывна. Иначе — копия. |
| **Транспонирование** | **View** | `b = a.T` | Меняются только метаданные (strides). |
| **Метод `.view()`** | **View** | `b = a.view()` | Явное создание представления. |
| **Метод `.copy()`** | **Copy** | `b = a.copy()` | Явное создание независимой копии. |
| **Продвинутая индексация** | **Copy** | `b = a[[0, 2]]` | Выборка по списку индексов всегда копирует. |
| **Булева маска** | **Copy** | `b = a[a > 0]` | Фильтрация всегда возвращает копию. |
| **Функции создания** | **Copy** | `b = np.sort(a)` | Большинство функций возвращают новый массив. |


## Эффективность по времени

```python
import math
import numpy as np
import time
import sys


python_list = [1, 2, 3, 4, 5]
numpy_array = np.array([1, 2, 3, 4, 5])

# ============================================================
# 1. СКОРОСТЬ ВЫПОЛНЕНИЯ ОПЕРАЦИЙ
# ============================================================

print("\n" + "=" * 60)
print("1. СРАВНЕНИЕ СКОРОСТИ ВЫПОЛНЕНИЯ ОПЕРАЦИЙ")
print("=" * 60)

# Размер тестовых данных
SIZE = 10_000_000

python_list = list(range(SIZE))
numpy_array = np.arange(SIZE)

# --- Тест 1: Умножение каждого элемента на 2 ---
print("\n📊 Тест 1: Умножение каждого элемента на 2")

start = time.time()
result_list = [x * 2 for x in python_list]
time_list = time.time() - start

start = time.time()
result_array = numpy_array * 2
time_array = time.time() - start

print(f"Python список:  {time_list:.4f} сек")
print(f"NumPy массив:   {time_array:.4f} сек")
print(f"Ускорение:      {time_list / time_array:.2f}x быстрее")

# --- Тест 2: Сложение двух массивов ---
print("\n📊 Тест 2: Сложение двух последовательностей")

python_list2 = list(range(SIZE))
numpy_array2 = np.arange(SIZE)

start = time.time()
result_list = [a + b for a, b in zip(python_list, python_list2)]
time_list = time.time() - start

start = time.time()
result_array = numpy_array + numpy_array2
time_array = time.time() - start

print(f"Python список:  {time_list:.4f} сек")
print(f"NumPy массив:   {time_array:.4f} сек")
print(f"Ускорение:      {time_list / time_array:.2f}x быстрее")

# --- Тест 3: Возведение в квадрат ---
print("\n📊 Тест 3: Возведение каждого элемента в квадрат")

start = time.time()
result_list = [x ** 2 for x in python_list]
time_list = time.time() - start

start = time.time()
result_array = numpy_array ** 2
time_array = time.time() - start

print(f"Python список:  {time_list:.4f} сек")
print(f"NumPy массив:   {time_array:.4f} сек")
print(f"Ускорение:      {time_list / time_array:.2f}x быстрее")

# --- Тест 4: Скалярное произведение (сумма произведений) ---
print("\n📊 Тест 4: Скалярное произведение (сумма произведений)")

start = time.time()
result_list = sum(a * b for a, b in zip(python_list, python_list2))
time_list = time.time() - start

start = time.time()
result_array = np.sum(numpy_array * numpy_array2)
time_array = time.time() - start

print(f"Python список:  {time_list:.4f} сек")
print(f"NumPy массив:   {time_array:.4f} сек")
print(f"Ускорение:      {time_list / time_array:.2f}x быстрее")

# --- Тест 5: Фильтрация элементов ---
print("\n📊 Тест 5: Фильтрация элементов (> 5000000)")

start = time.time()
result_list = [x for x in python_list if x > 5_000_000]
time_list = time.time() - start

start = time.time()
result_array = numpy_array[numpy_array > 5_000_000]
time_array = time.time() - start

print(f"Python список:  {time_list:.4f} сек")
print(f"NumPy массив:   {time_array:.4f} сек")
print(f"Ускорение:      {time_list / time_array:.2f}x быстрее")

# ============================================================
# 2. МАТЕМАТИЧЕСКИЕ ФУНКЦИИ
# ============================================================

print("\n" + "=" * 60)
print("2. МАТЕМАТИЧЕСКИЕ ФУНКЦИИ")
print("=" * 60)

numpy_array = np.random.rand(1_000_000)
python_list = numpy_array.tolist()

# --- Синус ---
print("\n📊 Тест: Вычисление синуса для 1 млн элементов")

start = time.time()
result_list = [math.sin(x) for x in python_list]
time_list = time.time() - start

start = time.time()
result_array = np.sin(numpy_array)
time_array = time.time() - start

print(f"Python список:  {time_list:.4f} сек")
print(f"NumPy массив:   {time_array:.4f} сек")
print(f"Ускорение:      {time_list / time_array:.2f}x быстрее")

# --- Статистика ---
print("\n📊 Тест: Вычисление среднего значения")

start = time.time()
result_list = sum(python_list) / len(python_list)
time_list = time.time() - start

start = time.time()
result_array = np.mean(numpy_array)
time_array = time.time() - start

print(f"Python список:  {time_list:.4f} сек")
print(f"NumPy массив:   {time_array:.4f} сек")
print(f"Ускорение:      {time_list / time_array:.2f}x быстрее")
```

# Задачи линейной алгебры

```python
import numpy as np

print("=== Задача 1: Базовые операции с матрицами ===")

# 1. Создание матриц 3x3
A = np.array([[1, 2, 3],
              [0, 1, 4],
              [5, 6, 0]], dtype=float)

B = np.array([[2, 0, 1],
              [3, 1, 0],
              [0, 2, 1]], dtype=float)

# 2. Транспонирование матрицы A (строки становятся столбцами)
A_T = A.T  # или np.transpose(A)
print(f"Матрица A:\n{A}")
print(f"Транспонированная A:\n{A_T}")

# 3. Матричное умножение (A * B)
# В numpy для матричного умножения лучше использовать @ или np.dot
C = A @ B
print(f"Результат умножения A @ B:\n{C}")

# 4. Вычисление определителя (Det)
det_A = np.linalg.det(A)
print(f"Определитель матрицы A: {det_A:.4f}")

# 5. Вычисление обратной матрицы (Inverse)
# Обратная матрица существует, только если определитель не равен 0
if not np.isclose(det_A, 0):
    A_inv = np.linalg.inv(A)
    print(f"Обратная матрица A^-1:\n{A_inv}")

    # Проверка: A * A^-1 должна давать единичную матрицу
    I = A @ A_inv
    print(f"Проверка (A * A^-1):\n{I}")
else:
    print("Матрица вырожденная, обратной не существует.")
```

```python
print("\n=== Задача 2: Решение системы линейных уравнений ===")

# Коэффициенты при неизвестных (Матрица A)
A = np.array([
    [2, 1, -1],
    [-3, -1, 2],
    [-2, 1, 2]
], dtype=float)

# Свободные члены (Вектор b)
b = np.array([8, -11, -3], dtype=float)

# Решение системы Ax = b
# np.linalg.solve использует метод LU-разложения под капотом
try:
    x = np.linalg.solve(A, b)
    print(f"Вектор решения x: {x}")

    # Проверка правильности решения (A @ x должно быть равно b)
    b_calc = A @ x
    print(f"Проверка (A @ x): {b_calc}")
    print(f"Исходный вектор b: {b}")

    # np.allclose проверяет равенство массивов с учетом погрешности float
    if np.allclose(b_calc, b):
        print("Решение верно!")
    else:
        print("Решение имеет погрешность.")

except np.linalg.LinAlgError:
    print("Система не имеет единственного решения (матрица вырожденная).")
```

```python
print("\n=== Задача 3: Собственные значения и векторы ===")

# Создадим симметричную матрицу 3x3 (у симметричных матриц вещественные собственные значения)
M = np.array([
    [4, 1, 1],
    [1, 4, 1],
    [1, 1, 4]
], dtype=float)

print(f"Матрица M:\n{M}")

# Вычисление собственных значений (eigenvalues) и векторов (eigenvectors)
eigenvalues, eigenvectors = np.linalg.eig(M)

print(f"Собственные значения (lambda): {eigenvalues}")
print(f"Собственные векторы (столбцы матрицы V):\n{eigenvectors}")

# ВЕРИФИКАЦИЯ: Проверим равенство M @ v = lambda * v для первого собственного вектора
v1 = eigenvectors[:, 0]  # Первый собственный вектор
l1 = eigenvalues[0]      # Первое собственное значение

left_side = M @ v1       # M * v
right_side = l1 * v1     # lambda * v

print(f"\nПроверка для 1-го собственного пары:")
print(f"M @ v1: {left_side}")
print(f"l1 * v1: {right_side}")

if np.allclose(left_side, right_side):
    print("Свойство собственных векторов подтверждено!")
```

Библиотека **pandas** поддерживает чтение данных из огромного количества форматов. Вот основной список, сгруппированный по типу данных, вместе с соответствующими функциями:

### 1. Текстовые файлы (самые популярные)
| Формат | Расширение | Функция pandas |
| :--- | :--- | :--- |
| **CSV** (Comma Separated Values) | `.csv` | `pd.read_csv()` |
| **TSV** (Tab Separated Values) | `.tsv`, `.txt` | `pd.read_csv(sep='\t')` или `pd.read_table()` |
| **JSON** | `.json` | `pd.read_json()` |
| **XML** | `.xml` | `pd.read_xml()` (доступно с версии 1.3.0) |
| **Фиксированная ширина** | `.txt`, `.fwf` | `pd.read_fwf()` |

### 2. Табличные процессоры
| Формат | Расширение | Функция pandas | Примечание |
| :--- | :--- | :--- | :--- |
| **Excel** | `.xls`, `.xlsx` | `pd.read_excel()` | Требует движок (`openpyxl` или `xlrd`) |
| **OpenDocument** | `.ods` | `pd.read_excel()` | Требует движок `odf` |

### 3. Бинарные и высокопроизводительные форматы
| Формат | Расширение | Функция pandas | Примечание |
| :--- | :--- | :--- | :--- |
| **Parquet** | `.parquet` | `pd.read_parquet()` | Требует `pyarrow` или `fastparquet` |
| **Feather** | `.feather` | `pd.read_feather()` | Требует `pyarrow` или `feather-format` |
| **HDF5** | `.h5`, `.hdf` | `pd.read_hdf()` | Требует `PyTables` |
| **Pickle** (сериализация Python) | `.pkl`, `.pickle` | `pd.read_pickle()` | Встроен в Python |

### 4. Базы данных (SQL)
| Формат | Функция pandas | Примечание |
| :--- | :--- | :--- |
| **SQL** (любая СУБД) | `pd.read_sql()`, `pd.read_sql_query()`, `pd.read_sql_table()` | Требует `SQLAlchemy` и драйвер БД (например, `psycopg2` для Postgres) |
| **Google BigQuery** | `pd.read_gbq()` | Требует библиотеку `pandas-gbq` |


### 5. Веб и буфер обмена
| Источник | Функция pandas | Примечание |
| :--- | :--- | :--- |
| **HTML таблицы** | `pd.read_html()` | Возвращает список датафреймов (по одному на таблицу) |
| **Буфер обмена** | `pd.read_clipboard()` | Копируете таблицу в Excel/браузере -> вставляете в код |

---

### Важные нюансы:
1.  **Зависимости:** Для некоторых форматов (Excel, Parquet, SQL, HDF5) нужно доустанавливать дополнительные библиотеки. Например:
    *   Для Excel: `pip install openpyxl`
    *   Для Parquet: `pip install pyarrow`
    *   Для SQL: `pip install sqlalchemy`
2.  **Запись:** Для каждого `read_...` почти всегда есть парная функция `to_...` (например, `df.to_csv()`, `df.to_excel()`), чтобы сохранить данные обратно.
3.  **Кодировка:** При чтении текстовых файлов (CSV, TXT) часто приходится указывать кодировку, например: `pd.read_csv('file.csv', encoding='utf-8')` или `encoding='cp1251'` (для русской Windows).

### all и any

any(): Возвращает True, если хотя бы один элемент истинный (логическое ИЛИ).

all(): Возвращает True, если все элементы истинные (логическое И).

```python
df = pd.DataFrame({
    'A': [1, 0, 5, 0],      # 0 считается False, остальные True
    'B': [True, True, False, True],
    'C': [10, 20, 30, 40],  # Все True (не нули)
    'D': [np.nan, 2, 3, 4]  # Есть пропуск
})

print("=== Исходный DataFrame ===")
print(df)
print("\n")

# ==========================================
# 1. Работа с Series (одна колонка)
# ==========================================
print("=== 1. Проверка Series (колонки) ===")

print(f"Колонка 'A' (есть ли не нули?): {df['A'].any()}")
print(f"Колонка 'A' (все не нули?):    {df['A'].all()}")
print(f"Колонка 'C' (все не нули?):    {df['C'].all()}")
print("\n")
```

```python
print("=== 2. Проверка DataFrame (axis) ===")

# axis=0 (по умолчанию) -> проверяем каждый СТОЛБЕЦ (спускаемся вниз)
print("any(axis=0) -> Есть ли True в каждом столбце?")
print(df.any())

print("\nall(axis=0) -> Все ли значения True в каждом столбце?")
print(df.all())

# axis=1 -> проверяем каждую СТРОКУ (идем вправо)
print("\nany(axis=1) -> Есть ли True в каждой строке?")
print(df.any(axis=1))

print("\nall(axis=1) -> Все ли значения True в каждой строке?")
print(df.all(axis=1))
print("\n")
```

## merge и concat

**pd.concat()** — это стекание (stacking). Вы просто склеиваете таблицы друг за другом (по строкам) или друг рядом с другом (по столбцам), ориентируясь на индексы. Это как сложить листы бумаги в стопку.

**pd.merge()** — это соединение (joining). Вы объединяете таблицы на основе общих значений в столбцах (ключей), подобно оператору JOIN в SQL. Это как связать две таблицы по общему ID.

```python
# Таблица 1: Сотрудники (Имя и ID)
df_employees = pd.DataFrame({
    'id': [1, 2, 3],
    'name': ['Alice', 'Bob', 'Charlie']
})

# Таблица 2: Зарплаты (ID и Зарплата)
df_salaries = pd.DataFrame({
    'id': [2, 3, 4],  # Обратите внимание: ID 1 нет, есть новый ID 4
    'salary': [5000, 6000, 7000]
})
```

```python
# Таблица 3: Новые сотрудники
df_new_employees = pd.DataFrame({
    'id': [4, 5],
    'name': ['David', 'Eva']
})

# Склеиваем вертикально
result_concat_vertical = pd.concat([df_employees, df_new_employees], ignore_index=True)

print("\n=== pd.concat (вертикально) ===")
print(result_concat_vertical)
```

```python
# Склеиваем горизонтально
result_concat_horizontal = pd.concat([df_employees, df_salaries], axis=1)

print("\n=== pd.concat (горизонтально) ===")
print(result_concat_horizontal)
```

```python
result_merge = pd.merge(df_employees, df_salaries, on='id', how='inner')

print("\n=== pd.merge (inner join по id) ===")
print(result_merge)
```

### Типы соединений в merge

```python
result_left = pd.merge(df_employees, df_salaries, on='id', how='left')
print("\n=== pd.merge (left join) ===")
print(result_left)
```

```python
result_left = pd.merge(df_employees, df_salaries, on='id', how='right')
print("\n=== pd.merge (left join) ===")
print(result_left)
```

#### Сравнение по времени с for

```python
import pandas as pd
import numpy as np
import time

# --- 1. Подготовка данных ---
# Создаем большой DataFrame, чтобы разница во времени была заметна
N_ROWS = 100_000
print(f"Генерация данных: {N_ROWS} строк...")
df = pd.DataFrame({
    'A': np.random.randint(1, 100, N_ROWS),
    'B': np.random.randint(1, 100, N_ROWS)
})

# Функция, которую будем применять
def complex_logic(a, b):
    if a > b:
        return a * 2 + b
    else:
        return a + b * 2

# --- 2. Функция для замера времени ---
def measure_time(method_name, func):
    start_time = time.perf_counter()
    func()
    end_time = time.perf_counter()
    duration = end_time - start_time
    print(f"{method_name:20} | Время: {duration:.4f} сек")
    return duration

# --- 3. Тестирование методов ---

results = {}


# МЕТОД: apply (axis=1)
# Применяем функцию к каждой строке
def test_apply():
    df['result_apply'] = df.apply(lambda row: complex_logic(row['A'], row['B']), axis=1)

results['Apply (axis=1)'] = measure_time("2. Apply (axis=1)", test_apply)

# МЕТОД: Цикл for + iterrows()
# Классический "медленный" цикл по строкам DataFrame
def test_for_iterrows():
    res = []
    for index, row in df.iterrows():
        res.append(complex_logic(row['A'], row['B']))
    df['result_for'] = res

results['For (iterrows)'] = measure_time("3. For (iterrows)", test_for_iterrows)


# --- 4. Анализ результатов ---
print("\n" + "="*40)
print("АНАЛИЗ ПРОИЗВОДИТЕЛЬНОСТИ")
print("="*40)

# Находим самое быстрое время (обычно это векторизация)
best_time = min(results.values())
best_method = [k for k, v in results.items() if v == best_time][0]

print(f"🏆 Лидер: {best_method}")

for method, t in results.items():
    if t == best_time:
        continue
    # Считаем, во сколько раз метод медленнее лидера
    ratio = t / best_time
    print(f"⚠️  {method} медленнее лидера в {ratio:.1f} раз")

print("="*40)
```

## Оптимизация памяти (Data Types)
Самый простой способ ускорить работу и уменьшить потребление RAM — использовать правильные типы данных. По умолчанию pandas часто использует «тяжелые» типы (например, int64, float64, object).
Если у вас числа от 0 до 100, нет смысла хранить их в int64 (занимает 8 байт). Достаточно int8 (1 байт).

```python
df = pd.DataFrame({
    'small_int': np.random.randint(0, 100, 100000),      # Займет int64
    'large_int': np.random.randint(0, 100000, 100000),   # Займет int64
    'float_col': np.random.rand(100000)                  # Займет float64
})

print("До оптимизации:")
print(df.info(memory_usage='deep'))

# Оптимизация
df['small_int'] = df['small_int'].astype('int8')       # 1 байт
df['float_col'] = df['float_col'].astype('float32')    # 4 байта вместо 8

print("\nПосле оптимизации:")
print(df.info(memory_usage='deep'))
```

Polars представляет собой dataframe-библиотеку с OLAP Query Engine, реализованным на языке программирования Rust с использованием колоночного формата Apache Arrow в качестве модели памяти.

В отличие от Pandas, который построен поверх NumPy и ограничен особенностями работы Python с памятью, Polars полностью контролирует все аспекты управления данными на системном уровне. Rust компилируется в машинный код без необходимости интерпретации, обеспечивает безопасную многопоточность и эффективное управление памятью без сборщика мусора.

Параллелизм в Polars реализован на фундаментальном уровне. В то время как Pandas работает в однопоточном режиме и требует дополнительных библиотек вроде Dask для распараллеливания, Polars автоматически использует все доступные ядра процессора для выполнения операций.

**Ленивые вычисления - основа быстродействия**

В то время как eager API работает подобно Pandas, выполняя операции сразу при вызове, lazy API позволяет построить граф вычислений, который затем оптимизируется перед выполнением.

**Существует несколько техник оптимизации запросов в lazy режиме:**

**Predicate Pushdown** — фильтры проталкиваются как можно ближе к источнику данных. Если вы читаете Parquet файл и применяете фильтр, Polars может прочитать только нужные row groups, даже не загружая остальные данные в память.

**Projection Pushdown** — если в конечном результате нужны только определенные колонки, Polars не будет загружать остальные колонки из источника данных, экономя память и время.

**Common Subexpression Elimination** — если одно и то же выражение используется несколько раз, оно будет вычислено только один раз.

**Query Reordering** — операции переставляются для минимизации объема обрабатываемых данных. Например, фильтрация будет выполнена до группировки, чтобы уменьшить количество групп.

## Сравнение по времени

Маленькие датасеты (< 10,000 строк): Pandas иногда может быть быстрее на простых операциях благодаря меньшему overhead на инициализацию. Разница обычно составляет миллисекунды и незаметна для пользователя.

Средние датасеты (10,000 - 1,000,000 строк): Polars начинает показывать преимущество в 2-5 раз на большинстве операций. Особенно заметна разница на группировках и join'ах.

Большие датасеты (> 1,000,000 строк): Преимущество Polars становится подавляющим — ускорение в 10-30 раз становится нормой. На датасетах в десятки миллионов строк Pandas может просто не справиться из-за ограничений памяти, в то время как Polars продолжает работать эффективно.

Очень большие датасеты (> RAM): Здесь Polars с streaming engine показывает уникальную способность работать с данными, превышающими объем оперативной памяти, что для Pandas просто невозможно без использования дополнительных инструментов вроде Dask.

**Итого, на маленьких датасетах (менее 10,000 строк) Pandas иногда может быть быстрее на простых операциях фильтрации благодаря меньшему overhead на инициализацию.**
